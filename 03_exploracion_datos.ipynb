```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('datos.csv')
df.head(5)
```

```python
df.describe()
```

```python

goles_local = df[df['equipo_local'] == "Cebollitas FC"]['goles_local'].mean()
goles_visitante = df[df['equipo_visitante'] == "Cebollitas FC"]['goles_visitante'].mean()   

print(f"Goles promedio como local: {goles_local}")
print(f"Goles promedio como visitante: {goles_visitante}")
```

```python

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(
    df["goles_local"],
    kde=True,
    ax=ax[0],
    bins=10,
)
ax[0].set_title("Distribuicion de goles locales")

sns.histplot(df["goles_visitante"], kde=True, ax=ax[1], bins=10, color="black")
ax[1].set_title("Distribucion de goles visitantes")

plt.show()
```

```python
sns.boxplot(x=df["goles_local"])
plt.title('Boxplot de goles locales')

plt.show()
```

```python
sns.scatterplot(x='posesion_local (%)', y='goles_local', data=df)
plt.title("Relacion posesion vs goles marcados de Equipo local")
```
