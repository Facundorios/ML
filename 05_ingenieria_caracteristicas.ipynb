```python
import pandas as pd

df = pd.read_csv('datos.csv')
df.head()
```

```python
# Diferencia de goles
df['diferencia_goles'] = df['goles_local'] - df['goles_visitante']

# Resultado
df[['goles_local', 'goles_visitante', 'diferencia_goles']].head()
```

```python
# Ratio de tiros al arco por porcentaje de posesión
df['ratio_tiros_posesion_local'] = df['tiros_arco_local'] / df['posesion_local (%)']

# Resultado
df[['tiros_arco_local', 'posesion_local (%)', 'ratio_tiros_posesion_local']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma de deiferencia de goles
sns.histplot(df['diferencia_goles'], bins=15,kde=True)
plt.title('Distribuciond de goles')
plt.xlabel('Diferencia de goles vissitabnte - local')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Analisis de correlaciones
corr_vars = ['goles_local', 'goles_visitante', 'diferencia_goles', 'ratio_tiros_posesion_local']
corr_matrix = df[corr_vars].corr()

# Mapa de calor
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlación entre variables originales y nuevas')
plt.show()
```
