```python
import pandas as pd


```

```python
# Cargar datos desde un archivo CSV
datos = pd.read_csv('datos.csv')
datos.head()
```

```python
# Revisa la estructura de los datos
datos.info()
```

```python
datos.isnull().sum()
datos['goles_local'] = datos['goles_local'].fillna(datos['goles_local'].mean())
print("Valores nulos después de la imputación:")
print(datos.isnull().sum())
```

```python
datos = pd.get_dummies(datos, columns=['equipo_local', 'equipo_visitante'])
datos.head()
```

```python
# Eliminar columnas duplicadas
datos.drop_duplicates(inplace=True)

print("Datos Antes de eliminar duplicados:", len(datos))
datos.drop_duplicates(inplace=True)
print("Datos después de eliminar duplicados:", len(datos))
```

```python
# Manejar fechas
datos['fecha_partido'] = pd.to_datetime(datos['fecha_partido'], errors='coerce')

print("Fechas invalidad NaT luego de la conversión")
print(datos['fecha_partido'].isnull().sum())
```

```python
# Resumen
print(datos.info())
```

```python
print("Columnas con nulos tras limpieza")
print(datos.isnull().sum())
```
